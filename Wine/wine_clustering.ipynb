# Usare il metodo di elbow per trovare il numero ottimale di cluster.

```python

from sklearn.cluster import KMeans

iner = []

for i in range(1, 20):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(wine)
    iner.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 20), iner, color='blue')
plt.scatter(range(1, 20), iner, color='red', marker='o')
plt.title('Metodo elbow')
plt.xlabel('Numero di cluster')
plt.ylabel('Inertia')
plt.xticks(range(1, 20))
plt.grid(True)
plt.show()
```

# K-means Clustering Model

```python
kmeans = KMeans(n_clusters=3, init='k-means++',random_state=42)
y = kmeans.fit_predict(wine)
print(y)
```

```python
plt.figure(figsize=(10, 6), dpi=80)
scatter = plt.scatter(wine[:, 0], wine[:, 1], c=y, s=50, cmap='Set1')

centers = kmeans.cluster_centers_

plt.scatter(centers[:, 0], centers[:, 1], c=np.unique(y), s=200, alpha=0.5, edgecolors='black', linewidths=2, cmap='Set1')

# Creazione della legenda
labels = ['Vino 1', 'Vino 2', 'Vino 3']
colors = ['red', 'grey', 'orange']
handles = [plt.Line2D([], [], marker='o', markersize=8, linestyle='None', color=color) for color in colors]
plt.legend(handles, labels)

plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(wine, y)
print("Silhouette Score:", silhouette_avg)
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Calcolo dei valori di silhouette per diversi numeri di cluster
silhouette_scores = []
inertia_values = []
num_clusters = range(2, 10)

for k in num_clusters:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    labels = kmeans.fit_predict(wine)
    silhouette_scores.append(silhouette_score(wine, labels))
    inertia_values.append(kmeans.inertia_)

# Grafico dei valori di silhouette
plt.figure(figsize=(10, 6))
plt.plot(num_clusters, silhouette_scores, marker='o')
plt.xlabel('Numero di cluster')
plt.ylabel('Valore di silhouette')
plt.title('Valore di silhouette per numero di cluster')
plt.xticks(num_clusters)
plt.grid(True)
plt.show()

# Grafico dei valori di inertia
plt.figure(figsize=(10, 6))
plt.plot(num_clusters, inertia_values, marker='o')
plt.xlabel('Numero di cluster')
plt.ylabel('Inertia')
plt.title('Inertia per numero di cluster')
plt.xticks(num_clusters)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Creazione dell'istanza del modello di clustering gerarchico
model = AgglomerativeClustering(n_clusters=3)

# Effettuare il clustering gerarchico sul dataset "wine"
labels = model.fit_predict(wine)

# Visualizzare i risultati del clustering
plt.figure(figsize=(10, 6))
plt.scatter(wine[:, 0], wine[:, 1], c=labels, s=50, cmap='viridis')
plt.xlabel('Prima caratteristica')
plt.ylabel('Seconda caratteristica')
plt.title('Clustering gerarchico su wine')

# Creazione della legenda
labels = ['Vino 1', 'Vino 2', 'Vino 3']
colors = ['darkblue', 'grey', 'yellow']
handles = [plt.Line2D([], [], marker='o', markersize=8, linestyle='None', color=color) for color in colors]
plt.legend(handles, labels)

plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Calcolo della matrice di linkage per il clustering gerarchico
linkage_matrix = linkage(wine, method='ward')

# Creazione del dendrogramma
plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix, color_threshold=0)

# Impostazioni dell'asse x
plt.xlabel('Indici dei campioni')
plt.ylabel('Distanza')

# Titolo del dendrogramma
plt.title('Dendrogramma del clustering gerarchico su wine')

plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Applicazione della PCA per ridurre la dimensionalità del dataset
pca = PCA(n_components=2)
wine_pca = pca.fit_transform(wine)

# Applicazione del clustering K-means sul dataset originale
kmeans_original = KMeans(n_clusters=3, random_state=42)
labels_original = kmeans_original.fit_predict(wine)
inertia_original = kmeans_original.inertia_
silhouette_original = silhouette_score(wine, labels_original)

# Applicazione del clustering K-means sul dataset PCA
kmeans_pca = KMeans(n_clusters=3, random_state=42)
labels_pca = kmeans_pca.fit_predict(wine_pca)
inertia_pca = kmeans_pca.inertia_
silhouette_pca = silhouette_score(wine_pca, labels_pca)

# Creazione del dataframe
data = {
    'Dataset': ['Originale', 'PCA'],
    'Inertia': [inertia_original, inertia_pca],
    'Silhouette': [silhouette_original, silhouette_pca]
}
df = pd.DataFrame(data)

# Stampa del dataframe
df
```

```python
import matplotlib.pyplot as plt

# Creazione del grafico
fig, axes = plt.subplots(2, 1, figsize=(8, 10))

# Grafico per l'inertia
axes[0].bar(df['Dataset'], df['Inertia'], color=['blue', 'orange'])
axes[0].set_title('Valori di Inertia')
axes[0].set_ylabel('Inertia')

# Grafico per la silhouette
axes[1].bar(df['Dataset'], df['Silhouette'], color=['blue', 'orange'])
axes[1].set_title('Valori di Silhouette')
axes[1].set_ylabel('Silhouette')

# Visualizzazione del grafico
plt.tight_layout()
plt.show()
```
